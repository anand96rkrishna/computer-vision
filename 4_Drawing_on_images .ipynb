```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#define imshow function
def imshow(title="image",image=None,size=10):
  w,h=image.shape[0],image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(size * aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show
```

Black square canvas

```python
image=np.zeros((600,600,3),np.uint8)
image_gray=np.zeros((600,600),np.uint8)
imshow("Black_canvas-RGB color",image)
imshow("Black canvas-Grayscale",image_gray)
```

Draw line in black square

```python
cv2.line(image,(0,0),(600,600),(255,127,0),10)
#..............starting co-ordinate,ending coordinate,color,thickess
imshow("black canvas with diagonal line",image)
```

Drawing Rectangle

```python
image=np.zeros((600,600,3),np.uint8)
cv2.rectangle(image,(100,100),(300,250),(255,127,0),10)

#..............starting co-ordinate,ending coordinate
imshow("black canvas with  rectangle",image)
```

Drawing circle


```python
image=np.zeros((600,600,3),np.uint8)
cv2.circle(image, (350,350),100,(255,127,20),1)

#..............starting co-ordinate,ending coordinate
imshow("black canvas with  circle",image)
```

Drawing Polygons


```python
image=np.zeros((512,512,3), np.uint8)
#define points
pts=np.array([[10,50], [400,50], [90,200], [50,500]],np.int32)
#now reshape our points in form required by polylines
pts=pts.reshape((-1,1,2))
cv2.polylines(image,[pts],True,(0,0,255),3)
imshow("Black Canvas with red polygpn,",image)
```

Adding Text

```python
image=np.zeros((1000,1000,3),np.uint8)
ourstring="Anandakrishnan"
cv2.putText(image,ourstring,(155,200),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,3,(40,200,0),4)
imshow("anandhu",image)
```
